### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
combined_sales = pd.concat([atletic_sales_2020_df, atletic_sales_2021_df], axis=0).reset_index(drop=True)

# Group by "region", "state", and "city" and sum the "products_sold" for each group
grouped_sales = combined_sales.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the sum column to "Total_Products_Sold"
grouped_sales = grouped_sales.rename(columns={'units_sold': 'Total_Products_Sold'})

# Sort the DataFrame by "Total_Products_Sold" in descending order and show the top 5 results
top_5_results = grouped_sales.sort_values(by='Total_Products_Sold', ascending=False).head(5)

# Display the top 5 results
print("Top 5 Results:")
top_5_results

                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
combined_sales = pd.concat([atletic_sales_2020_df, atletic_sales_2021_df], axis=0).reset_index(drop=True)

# Create a pivot table to show the number of products sold for region, state, and city
pivot_table = pd.pivot_table(combined_sales, values='units_sold', index=['region', 'state', 'city'], aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table = pivot_table.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results
top_5_results = pivot_table.sort_values(by='Total_Products_Sold', ascending=False).head(5)

# Display the top 5 results
print("Top 5 Results:")
top_5_results

```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
combined_sales = pd.concat([atletic_sales_2020_df, atletic_sales_2021_df], axis=0).reset_index(drop=True)

# Group by "region", "state", and "city" and sum the "total_sales" for each group
grouped_sales = combined_sales.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

# Rename the "total_sales" column to "Total Sales"
grouped_sales = grouped_sales.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results
top_5_results = grouped_sales.sort_values(by='Total Sales', ascending=False).head(5)

# Display the top 5 results
print("Top 5 Results:")
top_5_results
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
combined_sales = pd.concat([atletic_sales_2020_df, atletic_sales_2021_df], axis=0).reset_index(drop=True)

# Create a pivot table to show the total sales for each region, state, and city
pivot_table = pd.pivot_table(combined_sales, values='total_sales', index=['region', 'state', 'city'], aggfunc='sum').reset_index()

# Rename the "total_sales" column to "Total Sales"
pivot_table = pivot_table.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results
top_5_results = pivot_table.sort_values(by='Total Sales', ascending=False).head(5)

# Display the top 5 results
print("Top 5 Results:")
top_5_results
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
combined_sales = pd.concat([atletic_sales_2020_df, atletic_sales_2021_df], axis=0).reset_index(drop=True)   

women_footwear_sales = combined_sales[combined_sales['product'] == "Women's Athletic Footwear"]

# Display the women's athletic footwear sales data
print("Women's Athletic Footwear Sales Data:")
women_footwear_sales
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
combined_sales = pd.concat([atletic_sales_2020_df, atletic_sales_2021_df], axis=0).reset_index(drop=True)

# Filter the data for women's athletic footwear
women_footwear_sales = combined_sales[combined_sales['product'] == "Women's Athletic Footwear"]

# Group by "retailer", "region", "state", and "city" and sum the "units_sold" for each group
grouped_women_footwear_sales = women_footwear_sales.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
grouped_women_footwear_sales = grouped_women_footwear_sales.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results
top_5_results = grouped_women_footwear_sales.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)

# Display the top 5 results
print("Top 5 Women's Athletic Footwear Sales:")
top_5_results
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
combined_sales = pd.concat([atletic_sales_2020_df, atletic_sales_2021_df], axis=0).reset_index(drop=True)

# Filter the data for women's athletic footwear
women_footwear_sales = combined_sales[combined_sales['product'] == "Women's Athletic Footwear"]

# Create a pivot table to show the total number of women's athletic footwear sold for each retailer, region, state, and city
pivot_table = pd.pivot_table(women_footwear_sales,
                             values='units_sold',
                             index=['retailer', 'region', 'state', 'city'],
                             aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_table = pivot_table.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results
top_5_results = pivot_table.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)

# Display the top 5 results
print("Top 5 Women's Athletic Footwear Sales:")
top_5_results
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
combined_sales = pd.concat([atletic_sales_2020_df, atletic_sales_2021_df], axis=0).reset_index(drop=True)

# Create a pivot table with 'invoice_date' as the index and "total_sales" as the values
pivot_table = pd.pivot_table(combined_sales,
                             values='total_sales',
                             index='invoice_date',
                             aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table = pivot_table.rename(columns={'total_sales': 'Total Sales'})

# Show the pivot table
print("Pivot Table:")
pivot_table
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
combined_sales = pd.concat([atletic_sales_2020_df, atletic_sales_2021_df], axis=0).reset_index(drop=True)

pivot_table = pd.pivot_table(combined_sales,
                             values='total_sales',
                             index='invoice_date',
                             aggfunc='sum').reset_index()

pivot_table['invoice_date'] = pd.to_datetime(pivot_table['invoice_date'])

resampled_pivot = pivot_table.resample('D', on='invoice_date').sum()

# Display the resampled pivot table
print("Resampled Pivot Table:")
resampled_pivot
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
# Create the original pivot table with 'invoice_date' as the index and "total_sales" as the values
pivot_table = pd.pivot_table(combined_sales,
                             values='total_sales',
                             index='invoice_date',
                             aggfunc='sum').reset_index()

# Ensure "invoice_date" is in datetime format
pivot_table['invoice_date'] = pd.to_datetime(pivot_table['invoice_date'])

# Resample the pivot table into weekly bins and get the total sales for each week
resampled_pivot_weekly = pivot_table.resample('W-Mon', on='invoice_date').sum()

# Sort the resampled pivot table in ascending order on "Total Sales"
sorted_resampled_pivot_weekly = resampled_pivot_weekly.sort_values(by='total_sales', ascending=True)

# Display the sorted resampled pivot table for weekly bins
print("Sorted Resampled Pivot Table (Weekly Bins):")
sorted_resampled_pivot_weekly
```
